## Vizualization

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_palette("Paired")
# set plot style: grey grid in the background:
sns.set(style="white")

# Set the figure size
plt.figure(figsize=(5, 5))

# plot a bar chart
# Create a count plot using Seaborn
ax = sns.countplot(y='Q1',
                   data=cust_survey,
                   linewidth=1,
                   edgecolor='black',
                   palette='Paired')
plt.title('Percentage of Customers Celebrating Women History Month\n',
          fontname='Times New Roman',
          fontsize=14,
          fontweight='light')
plt.ylabel('Planning to celebrate WHM\n',
           fontname='Times New Roman',
           fontsize=12,
           fontweight='light')
ax.set_yticklabels(ax.get_yticklabels(),
                   fontname='Times New Roman',
                   fontsize=10)
#remove x-axis label
plt.xlabel(' ')  #create code to remove tile

# remove the spines
sns.despine()

# remove the x-axis line
sns.despine(bottom=True)
# Remove x-axis
ax.set(xticks=[])

#loop through each bar in the countplot and add a text annotation with the percentage value
total = len(cust_survey['Q1'])
palette = sns.color_palette("Paired")  # define color palette

for i, p in enumerate(ax.patches):
    percentage = '    {:.1f}%'.format(100 * p.get_width() / total)
    x = p.get_x() + p.get_width() + 0.02
    y = p.get_y() + p.get_height() / 2
    ax.annotate(percentage, (x, y),
                color=palette[i % len(palette)],
                fontweight="bold")

#View
plt.show()
```
